###      1.1.1. display the quantized image in (4,8,16,32,64,128) and calculate their MSE

```python
#reading the image
lena = cv2.imread('lena.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
def quantize(image, n_bits):
    coeff = 2**8 // 2**n_bits
    coeff2= 2**8 // (2**n_bits - 1)
    return (image // coeff) * coeff2
```

```python
flat=lena.flatten()
```

```python
plt.subplot(121)
plt.imshow(lena)
plt.title('original image')

#convert to grayscale
plt.set_cmap('gray')
plt.show()
```

```python
img128levels = quantize(lena,7)
plt.subplot(121)
plt.imshow(img128levels)
plt.title('128 levels')

img_arr128 = np.array(lena)
#MSE
mse = np.mean((img_arr128 - img128levels) ** 2)
print(f"Quantization level: {128}, MSE: {mse}")


plt.show()
```

###  1.1.3. By using point processing create new images to make it Darken, Lighten, Lower contrast, raised contrast and finally Invert image and display the results on gray scale Lena.

```python
img = cv2.imread('lena.bmp', cv2.IMREAD_GRAYSCALE)
```

##### original image 

```python
plt.subplot(121)
plt.imshow(lena)
plt.title('original grayscaled image')

data = np.asarray(lena)

#convert to grayscale
plt.set_cmap('gray')
plt.show()
```

#### Darken

```python
def darken(pixel_value):
    return pixel_value * 0.5

height, width = img.shape
dark_img = np.zeros((height, width), dtype=np.uint8)

dark_img = np.vectorize(darken)(img)
IM = Image.fromarray(dark_img)
plt.imshow(IM) 
plt.show()
```

####  lighten

```python
def lighten(pixel):
    return pixel * 1.5

height, width = img.shape
light_img = np.zeros((height, width), dtype=np.uint8)

light_img = np.vectorize(lighten)(img)
IM = Image.fromarray(light_img)
plt.imshow(IM) 
plt.show()
```

#### lower contrast

```python
def lower_contrast(pixel):
    return (pixel - np.mean(img)) * 0.5 + np.mean(img)

height, width = img.shape
low_contr_img = np.zeros((height, width), dtype=np.uint8)

low_contr_img = np.vectorize(lower_contrast)(img)
IM = Image.fromarray(low_contr_img)
plt.imshow(IM) 
plt.show()
```

#### raised contrast

```python
def raised_contrast(pixel):
    return (pixel - np.mean(img)) * 1.5 + np.mean(img)


height, width = img.shape
high_contr_img = np.zeros((height, width), dtype=np.uint8)

high_contr_img= np.vectorize(raised_contrast)(img)
IM = Image.fromarray(high_contr_img)
plt.imshow(IM) 
plt.show()
```

#### invert

```python
def invert(pixel):
    return 255 - pixel

height, width = img.shape
invert_img = np.zeros((height, width), dtype=np.uint8)

invert_img= np.vectorize(invert)(img)
IM = Image.fromarray(invert_img)
plt.imshow(IM) 
plt.show()
```
